```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('Books.csv')
books
```

```python
rating = pd.read_csv('Ratings.csv')
rating
```

```python
rating.drop(index=rating[rating["Book-Rating"]==0].index, inplace = True)
rating
```

```python
rating_count=pd.DataFrame(rating.groupby('ISBN').count())
rating_value=pd.DataFrame(rating.groupby('ISBN').mean())
```

```python
rating_value.drop(columns='User-ID',inplace=True )
rating_count.drop(columns='User-ID',inplace=True)
```

```python
rating_count.rename(columns={'Book-Rating':'Review Count'},inplace=True)
rating_value.rename(columns={'Book-Rating':'Average Rating'},inplace=True)
```

```python
books.drop(['Image-URL-S','Image-URL-M'],axis=1,inplace=True)
books.rename(columns={'Image-URL-L':'Cover Page'}, inplace=True)
```

```python
rating_count
```

```python
rating_value
```

```python
x =  rating.groupby('User-ID').count()['Book-Rating']>10
x
```

```python
knowledgable_user = x[x].index
knowledgable_user.size
```

```python
filtered_rating = rating[rating['User-ID'].isin(knowledgable_user)]
filtered_rating
```

```python
x=filtered_rating.groupby('ISBN').count()['User-ID']>10
x
```

```python
known_books= x[x].index
known_books
```

```python
filtered_rating=filtered_rating[filtered_rating['ISBN'].isin(known_books)]
filtered_rating
```

```python
filtered_data = books.merge(filtered_rating, how='right', on='ISBN')
```

```python
filtered_data.dropna(subset=['Book-Title'],inplace=True)
filtered_data.head()
```

```python
p_table = filtered_data.pivot_table(index=['User-ID'],columns=['Book-Title'],values=['Book-Rating'])
p_table
```

```python
p_table[('Book-Rating', '10 Lb. Penalty')]
```

```python
def show_image(val):
    return '<a href="{}"><img src="{}" width=50></img></a>'.format(val, val)
```

```python
def item_based_recommendation(book_title):
    book_title=str(book_title)
    
    if book_title in filtered_data['Book-Title'].values:
        book_rating = p_table[('Book-Rating', book_title)]
        recommendation_df=pd.DataFrame(p_table.corrwith(book_rating).sort_values(ascending=False)).reset_index(drop=False)  
        if book_title in [title for title in recommendation_df["Book-Title"]]:
            recommendation_df=recommendation_df.drop(recommendation_df[recommendation_df["Book-Title"]==book_title].index[0])
        
        low_rated=[]
        for i in recommendation_df["Book-Title"]:
            if filtered_data[filtered_data["Book-Title"]==i]["Book-Rating"].any() < 5:
                low_rated.append(i)
            if recommendation_df.shape[0] - len(low_rated) > 5:
                recommendation_df=recommendation_df[~recommendation_df["Book-Title"].isin(low_rated)]
            recommendations = filtered_data[filtered_data['Book-Title'].isin(recommendation_df['Book-Title'])]
        recommendations = recommendations.merge(rating_value, on=['ISBN'], how='left')         
        recommendations.sort_values(by=['Average Rating'],ascending=False,inplace=True)
        
        recommendations.dropna(subset=['Book-Title'],inplace=True)
        recommendations.drop_duplicates(subset=['Book-Title'],inplace=True)
        top_recommendation =recommendations[0:10]
        top_recommendation = top_recommendation.merge(rating_count, on='ISBN', how='left').sort_values(by=['Review Count'], ascending=False)
        
        
        
    else:
        print("Oh! That one must be a rare book. Sorry no recommendations for it")
        print("Would you like to try our top books:")
        recommendations = rating_count.merge(filtered_data, how='left',on='ISBN')
        recommendations.drop_duplicates(subset=['Book-Title'],inplace=True)
        recommendations.dropna(subset=['Book-Title'],inplace=True)
        recommendations.sort_values(by='Review Count',inplace=True, ascending=False)
        top_recommendation = recommendations[0:10]
        top_recommendation = top_recommendation.merge(rating_value, how='left',on='ISBN')

        top_recommendation.sort_values(by='Average Rating',inplace=True, ascending=False)
        
    top_recommendation.drop(columns=['User-ID', 'Book-Rating','Average Rating','Review Count'],inplace=True)   
    return top_recommendation.style.format({'Cover Page':show_image})
```

```python
item_based_recommendation('Harry Potter and the Goblet of Fire (Book 4)')
```

```python
item_based_recommendation('Alchemist')
```
